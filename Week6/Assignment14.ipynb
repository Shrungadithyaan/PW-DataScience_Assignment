### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight web framework for Python. It is designed to be simple, flexible, and easy to use, making it a popular choice for building web applications and APIs. Flask provides the basic tools and features needed to create web applications, while allowing developers to customize and extend its functionality as needed.

The advantages of Flask framework :

1. Simplicity: Flask has a simple and intuitive syntax, making it easy to learn and use for building web applications.

2. Flexibility: Flask follows a "micro" framework approach, allowing developers to choose and integrate additional libraries and components as needed.

3. Lightweight: Flask has minimal dependencies and is lightweight, resulting in efficient and fast performance.

4. Extensibility: Flask provides a modular design that allows developers to easily add or remove functionality using its rich ecosystem of extensions.

5. Testing and Debugging: Flask has built-in support for testing and debugging, making it easier to write and run tests and handle errors during development.

6. Active Community: Flask has a vibrant community of developers, providing resources, tutorials, and libraries, and ensuring continuous improvement and support for the framework.

7. Documentation: Flask offers comprehensive and well-documented official documentation, providing clear explanations and examples for easier understanding and utilization.
