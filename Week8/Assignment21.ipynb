## Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import plotly.express as px
import seaborn as sns
```

```python
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data.head()
```

```python
fig = px.scatter(titanic_data, x="age", y="fare")

fig.update_layout(title="Scatter Plot of Age vs. Fare in Titanic Dataset",
                  xaxis_title="Age",
                  yaxis_title="Fare")
fig.show()
```

## Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px
```

```python
tips_data = px.data.tips()
```

```python

tips_data.head()
```

```python
fig = px.box(tips_data, x="day", y="total_bill", color="sex", notched=True)

#adding titles and x&y axis labels
fig.update_layout(title="Box Plot of Total Bill by Day and Sex",
                  xaxis_title="Day",
                  yaxis_title="Total Bill")

fig.show()
```

## Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.express as px
df = px.data.tips()
df.head()
```

```python
fig = px.histogram(df, x= "sex" ,y="total_bill",pattern_shape="smoker",color="day")
fig.update_layout(title="Histogram of Sex by Total Bill",
                  xaxis_title="Sex",
                  yaxis_title="Total Bill")
fig.show()
```

## Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

```python
import plotly.express as px
```

```python
df = px.data.iris()
df.head()
```

```python
fig = px.scatter_matrix(df,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],color="species")
fig.update_layout(
    width=1200,
    height=800
)
fig.show()
```

## Q5. What is Distplot? Using Plotly express, plot a distplot.

A Distplot or distribution plot, depicts the variation in the data distribution

In Plotly Express, the distplot function is used to create a distribution plot, which is also known as a histogram with a kernel density estimate (KDE) overlaid on top. A distribution plot provides a visual representation of the distribution of a single numerical variable

```python
import plotly.express as px
df = px.data.tips()
df.head()
```

```python
fig = px.histogram(df, x="total_bill", y="tip", color="sex",
                   marginal="rug", # or violin, rug
                   hover_data=df.columns)

fig.show()
```

```python
#another approach of using distplot
import numpy as np
import plotly.figure_factory as ff
hist_data = [np.random.randn(100)]
labels = ['distplot']
figure = ff.create_distplot(hist_data, labels)
figure.show()
```
