The function is defined as

PMF(X) = P(X = x),

## Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

The Cumulative Density Function (CDF) is a function used to describe the cumulative probability distribution of a random variable. For both discrete and continuous random variables, the CDF provides the probability that the random variable takes on a value less than or equal to a specified value.

The CDF is used for several purposes in probability and statistics:

Calculation of Probabilities: The CDF provides an easy way to calculate probabilities for specific intervals. For example, P(a ≤ X ≤ b) is simply F(b) - F(a).

Determining Percentiles: Percentiles represent specific points in a distribution, such as the median (50th percentile) or quartiles (25th and 75th percentiles). These percentiles can be easily determined using the CDF.

Hypothesis Testing: The CDF is useful for hypothesis testing and calculating critical values.

Generating Random Samples: In some cases, random samples can be generated using the inverse CDF method, also known as the quantile function.

## Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

The normal distribution / Gaussian distribution is often applied in situations where the data exhibit a bell-shaped, symmetric pattern around the mean.

Some examples of situations where the normal distribution might be used as a model include:

Height and Weight of Individuals: The distribution of adult heights and weights tends to follow a roughly normal distribution, with most individuals close to the mean height and weight.

Test Scores: In educational testing, the scores of large groups of students on standardized tests often approximate a normal distribution.

Errors in Measurements: When measuring physical quantities with inherent randomness, such as errors in laboratory measurements or electronic noise, the distribution of errors is often modeled as normal.

Natural Phenomena: Many natural phenomena, such as the distribution of IQ scores, blood pressure levels, and reaction times, approximate a normal distribution.

The normal distribution is characterized by two parameters:

mean (μ)
standard deviation (σ).
Mean (μ):
The mean represents the central value around which the data are centered. It is the location parameter, determining the center of the distribution. Changing the mean shifts the entire distribution to the left or right without changing its shape.

Standard Deviation (σ):
The standard deviation is a measure of the spread or dispersion of the data points around the mean. It is the scale parameter, influencing the width of the distribution. A smaller standard deviation results in a narrower distribution, while a larger standard deviation results in a broader distribution.

Shape Change with Respect to the Two Paramaters
The shape of the normal distribution remains constant for a given mean, but the spread changes as the standard deviation varies. A smaller standard deviation makes the distribution more peaked and concentrated around the mean, while a larger standard deviation results in a flatter, wider distribution.

When the standard deviation is equal to 1, and the mean is equal to 0, the normal distribution is called the standard normal distribution. This standardized form allows us to compare and convert any normal distribution to the standard normal distribution using z-scores.

## Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.

The importance of Normal Distribution lies in the following aspects:

Central Limit Theorem: One of the key reasons for the significance of the normal distribution is the Central Limit Theorem (CLT). The CLT states that the sum (or average) of a large number of independent and identically distributed random variables, regardless of their original distribution, will be approximately normally distributed. This property makes the normal distribution a fundamental tool for inferential statistics and hypothesis testing.

Well-Understood Properties: The normal distribution is mathematically well-defined and analytically tractable. It allows for straightforward calculations and provides easily interpretable measures of central tendency (mean, median, mode) and variability (standard deviation).

Approximation for Many Phenomena: Although few real-world data are perfectly normally distributed, many natural processes and social phenomena tend to follow an approximately normal distribution, especially when the sample size is large. The normal distribution serves as a useful approximation in such cases.

Statistical Inference: The normal distribution is extensively used in statistical inference, including confidence interval estimation, hypothesis testing, and parameter estimation. It forms the basis for several statistical tests, such as t-tests and Z-tests.

Examples of Real-Life Normal Distributions:

IQ Scores: IQ (intelligence quotient) scores of a large population tend to follow an approximately normal distribution. The mean IQ is set at 100, and the standard deviation is typically set at 15. This distribution is useful for understanding intelligence levels in the population.

Height of Adult Humans: The heights of adult humans in a given population often approximate a normal distribution. The distribution is centered around the average height, with a standard deviation reflecting the variability of heights.

Exam Scores: In educational testing, the distribution of scores on standardized exams, especially when the exam is well-designed and covers a wide range of abilities, can closely follow a normal distribution.

Measurement Errors: In many scientific experiments, measurement errors are often assumed to follow a normal distribution. These errors can arise from various sources, such as instrument inaccuracies or observer biases.

Reaction Times: The time taken by individuals to respond to specific stimuli, known as reaction times, often exhibit a normal distribution when measured over a large sample of participants.

Stock Market Returns: The daily returns of well-established and stable stocks in the financial markets often exhibit a nearly normal distribution due to the complex interplay of numerous factors affecting stock prices.

## Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

The Bernoulli distribution is a discrete probability distribution that models a single binary outcome in a single trial of an experiment. It represents a situation where there are only two possible outcomes: success (usually denoted as "1") with probability "p" and failure (usually denoted as "0") with probability "q," where "q = 1 - p."

The Bernoulli distribution is named after the Swiss mathematician Jacob Bernoulli.

The Probability Mass Function (PMF) of the Bernoulli distribution is given by:

P(X = x) = p^x * q^(1-x)

where:

P(X = x) is the probability of observing outcome "x" (either 0 or 1).
"x" takes the value 1 for success and 0 for failure.
"p" is the probability of success.
"q" is the probability of failure.
For Example Consider flipping a fair coin:

The random variable X represents the outcome of the coin flip, where heads is a success (X = 1) and tails is a failure (X = 0). Since the coin is fair, the probability of getting heads (success) is 0.5, and the probability of getting tails (failure) is also 0.5.

The PMF of the Bernoulli distribution for this coin flip is as follows:

P(X = 1) = 0.5 (probability of getting heads)
P(X = 0) = 0.5 (probability of getting tails)
